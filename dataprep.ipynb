```python
# Importing Libraries
# These libraries will help us to move content from one directory to another

import os
import shutil
```

```python
Emotions = ['anger', 'disgust', 'fear', 'happiness', 'neutral', 'sadness','surprise']  # We have total 7 emotions 

# Making directory for each emotion
for e in Emotions :
    path = "/content/drive/MyDrive/Mini Project 5th Semester/Dataset/" + e
    os.mkdir(path)
```

```python
dirs = os.listdir("/content/drive/MyDrive/Mini Project 5th Semester/AudioData")
os.mkdir("/content/drive/MyDrive/Mini Project 5th Semester/Dataset/temp")
tempdir = "/content/drive/MyDrive/Mini Project 5th Semester/Dataset/temp/"

counters = [0, 0, 0, 0, 0, 0, 0]

# Dividing data from original dataset to different directory of emotions
for d in dirs :

    src = "/content/drive/MyDrive/Mini Project 5th Semester/AudioData/" + d + '/'
    dest = "/content/drive/MyDrive/Mini Project 5th Semester/Dataset/"
    files = os.listdir("/content/drive/MyDrive/Mini Project 5th Semester/AudioData/" + d)

    for f in files :
        if f[0] == 'a' :
            counters[0] = counters[0]+1
            shutil.copy(src + f, tempdir)
            os.rename(tempdir + f, tempdir + f[0] + str(counters[0]) + '.wav')
            shutil.move(tempdir + os.listdir(tempdir)[0] , dest + "anger/")


        elif f[0] == 'd' :
            counters[1] = counters[1]+1
            shutil.copy(src + f, tempdir)
            os.rename(tempdir + f, tempdir + f[0] + str(counters[1]) + '.wav')
            shutil.move(tempdir + os.listdir(tempdir)[0] , dest + "disgust/")

        elif f[0] == 'f' :
            counters[2] = counters[2]+1
            shutil.copy(src + f, tempdir)
            os.rename(tempdir + f, tempdir + f[0] + str(counters[2]) + '.wav')
            shutil.move(tempdir + os.listdir(tempdir)[0] , dest + "fear/")

        elif f[0] == 'h' :
            counters[3] = counters[3]+1
            shutil.copy(src + f, tempdir)
            os.rename(tempdir + f, tempdir + f[0] + str(counters[3]) + '.wav')
            shutil.move(tempdir + os.listdir(tempdir)[0] , dest + "happiness/")

        elif f[0] == 'n' :
            counters[4] = counters[4]+1
            shutil.copy(src + f, tempdir)
            os.rename(tempdir + f, tempdir + f[0] + str(counters[4]) + '.wav')
            shutil.move(tempdir + os.listdir(tempdir)[0] , dest + "neutral/")

        elif f[:2] == 'sa' :
            counters[5] = counters[5]+1
            shutil.copy(src + f, tempdir)
            os.rename(tempdir + f, tempdir + f[:2] + str(counters[5]) + '.wav')
            shutil.move(tempdir + os.listdir(tempdir)[0] , dest + "sadness/")

        elif f[:2] == 'su' :
            counters[6] = counters[6]+1
            shutil.copy(src + f, tempdir)
            os.rename(tempdir + f, tempdir + f[:2] + str(counters[6]) + '.wav')
            shutil.move(tempdir + os.listdir(tempdir)[0] , dest + "surprise/")

os.rmdir(tempdir)
```
