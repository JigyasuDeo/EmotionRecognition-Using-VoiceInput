```python
import matplotlib.pyplot as plot
from scipy.io import wavfile

samplingFrequency, signalData = wavfile.read('/content/drive/MyDrive/Mini Project 5th Semester/Dataset/anger/a1.wav')

plot.subplot(211)
plot.title('Angry')

plot.plot(signalData)
plot.xlabel('Sample')
plot.ylabel('Volume')

 

plot.subplot(212)

plot.specgram(signalData, mode='psd',Fs=samplingFrequency, cmap = 'magma')
plot.xlabel('Time')
plot.ylabel('Hz')

plot.show()
```
